```python
df = pd.read_csv("application_train.csv")
df.shape[0]
```

```python
missing = df.isna().sum().sort_values(ascending=False)
pd.DataFrame({"nb_nan": missing, "pct_nan": (missing/len(df)).round(3)}).head(50)
```

17 variables présentent un taux de valeurs manquantes supérieur à 60%.
Elles doivent certainement correspondre à des informations facultatives qui n'ont pas été transmises par les clients. 
La plupart de ces variables sont liées au logement, j'aurais donc tendance à dire qu'ici elles sont facultatives pour notre EDA.
Je vais donc drop toutes les colonnes qui dépassent ce seuil de 60%.


```python
threshold = 0.60
to_drop = [col for col in df.columns if df[col].isna().mean() > threshold]
print(f"{len(to_drop)} colonnes supprimées sur {len(df.columns)}")
dropped = df.drop(columns=to_drop)
dropped.shape
```

Les 17 colonnes ont été retirées du dataframe principal et stockées à part dans "dropped"
On va maintenant pouvoir débuter l'analyse univariée. 

```python
desc = df.describe().T
desc["missing_rate"] = (df.isna().sum() / len(df) * 100).round(2)
desc["skew"] = df.skew(numeric_only=True).round(2)
desc["kurtosis"] = df.kurtosis(numeric_only=True).round(2)
desc.head(15)
```
